```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
yen_df = pd.read_csv("C:\\Users\\willu\\Documents\\Yen-for-future-LinearRegression\\Starter_Code\\yen.csv")
```

```python
yen_df.head(10)
```

```python
yen_df = yen_df.set_index('Date')
```

```python
yen_df
```

```python
yen_df = yen_df.loc['1990-01-01':, :]
yen_df.head(10)
```

```python
yen_df['Returns'] = yen_df.Settle.pct_change()*100
```

```python
yen_df
```

```python
yen_df= yen_df.dropna()
```

```python
yen_df['lagged_returns'] = yen_df.Returns.shift()
yen_df = yen_df.dropna()
```

```python
yen_df.head()
```

```python
# Train test split 
train= yen_df[:'2017']
test = yen_df['2018':]
```

```python
X_train = train['lagged_returns'].to_frame()
X_test = test['lagged_returns'].to_frame()
y_train = train['Returns']
y_test= test['Returns']
```

```python
X_train
```

```python
# Creating a linear regression model from the data train/test split
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# making predictions
predict = model.predict(X_test)
```

```python
results = y_test.to_frame()
results['Predicted_Returns']= predict
```
